```python
#remove Na (null) Values
df = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, 'Batmobile', 'Bullwhip'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df
```

```python
df.dropna()
```

```python
#how method
df.dropna(how='all')
```

```python
df1 = pd.DataFrame({"name": ['Alfred', 'Batman', 'Catwoman'],
                   "toy": [np.nan, np.nan, np.nan],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df1
```

```python
df1.dropna(how='all', axis=1)
```

```python
# remove duplicates
df1 = pd.DataFrame({"name": ['Alfred', 'Batman', 'Alfred'],
                   "toy": ['car', 'batamobile', 'car'],
                   "born": [pd.NaT, pd.Timestamp("1940-04-25"),
                            pd.NaT]})
```

```python
df1
```

```python
df1.drop_duplicates(subset=['name'])
```

```python
df1.drop_duplicates(subset=['name'], keep='first')
```

```python
df1.drop_duplicates(subset=['name'], keep='last')
```

```python
df1.drop_duplicates(subset=['name'], keep=False)
```
